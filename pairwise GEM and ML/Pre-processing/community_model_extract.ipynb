```python
from zipfile import ZipFile
import glob
import pandas as pd
```

```python
destination = r"\community_extracted_exRxns\Healthy\\"

def extract_exRxn(zipped_folder, ends_with = '_rawExRxns.csv'):
    with ZipFile(zipped_folder, 'r') as zipObj:
        # Get a list of all archived file names from the zip
        listOfFileNames = zipObj.namelist()
        # Iterate over the file names
        for fileName in listOfFileNames:
            # Check filename endswith csv
            #if fileName.endswith(ends_with):
                # print(fileName)
                # Extract a single file from zip
            zipObj.extract(fileName, destination)
                
```

```python
exRxn_path = r"\community_extracted_exRxns\healthy\*\*_rawExRxns.csv"
tsv_path = r"\community_extracted_exRxns\healthy\*\*.tsv"
ex_rxn_files = glob.glob(exRxn_path)
rxn_name_files = glob.glob(tsv_path)
print(len(rxn_name_files))
healthy_sample_ex_rxns = pd.DataFrame()
for i in range(len(ex_rxn_files)):
    ex_rxn_df = pd.read_csv(ex_rxn_files[i], index_col=0).T
    rxn_names_df = pd.read_csv(rxn_name_files[i], sep="\t", index_col=0)
    # print(len(ex_rxn_df.index), "\t", len(rxn_names_df.index))
    if (len(ex_rxn_df.index) == len(rxn_names_df.index)):
        # print(rxn_names_df.index)
        ex_rxn_df.index = rxn_names_df.index
        ex_rxn_df['interaction_type'] = rxn_names_df[' TypeOfInteraction']
        healthy_sample_ex_rxns = pd.concat([healthy_sample_ex_rxns, ex_rxn_df], axis=0, ignore_index=False)
healthy_sample_ex_rxns.dropna(axis=1, inplace=True)
healthy_sample_ex_rxns.to_csv(r"community_extracted_exRxns\community_models_healthy.csv")
```

```python
exRxn_path = r"\community_extracted_exRxns\COVID\*\*_rawExRxns.csv"
tsv_path = r"\community_extracted_exRxns\COVID\*\*.tsv"
ex_rxn_files = glob.glob(exRxn_path)
rxn_name_files = glob.glob(tsv_path)
print(len(rxn_name_files))
COVID_sample_ex_rxns = pd.DataFrame()
for i in range(len(ex_rxn_files)):
    ex_rxn_df = pd.read_csv(ex_rxn_files[i], index_col=0).T
    rxn_names_df = pd.read_csv(rxn_name_files[i], sep="\t", index_col=0)
    # print(len(ex_rxn_df.index), "\t", len(rxn_names_df.index))
    if (len(ex_rxn_df.index) == len(rxn_names_df.index)):
        # print(rxn_names_df.index)
        ex_rxn_df.index = rxn_names_df.index
        ex_rxn_df['interaction_type'] = rxn_names_df[' TypeOfInteraction']
        COVID_sample_ex_rxns = pd.concat([COVID_sample_ex_rxns, ex_rxn_df], axis=0, ignore_index=False)
COVID_sample_ex_rxns.dropna(axis=1, inplace=True)
COVID_sample_ex_rxns.to_csv(r"community_extracted_exRxns\community_models_COVID.csv")
```

```python
healthy_sample_ex_rxns['Patient'] = "Healthy"
COVID_sample_ex_rxns['Patient'] = "COVID"

combined_samples = pd.concat([healthy_sample_ex_rxns, COVID_sample_ex_rxns], axis=0,  ignore_index=False)
combined_samples.dropna(axis=1, inplace=True)
combined_samples.to_csv(r"community_extracted_exRxns\community_samples_all.csv")
combined_samples.shape
```
